```python
import cv2
```

```python
#Using OpenCV to load an image
vic = cv2.imread(r'C:\Users\USER\Desktop\10alytics\cv\vic.jpg')
vic

#this instead automatically bring in the image as bunch of numbers
```

```python
type(vic)
```

```python
#OpenCV uses BGR instead of RGB
plt.imshow(vic)
```

```python
# Convert picture or iage from BGR to RGB

vicky = cv2.cvtColor(vic, cv2.COLOR_BGR2RGB)
plt.imshow(vicky)
```

```python
#Resizing the image - (Width and Height)
plt.imshow(cv2.resize(vicky, (750, 750)))
```

```python
#Saving the resized image on desktop
img = cv2.resize(vicky, (750,750))

#cONVERT IMG FROM BGR 2 RGB
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Export the image
cv2.imwrite(r"C:\Users\USER\Desktop\10alytics\cv\vicky.jpg", img1)
```

```python
#Resize Through percentage (%)
width = 0.5
height = 0.5

vicko = cv2.resize(vicky, (0,0), vicky, width, height)
plt.imshow(vicko)
```

```python
#Flipping the image (0 - Flips Horizontal axis)
vetim = cv2.flip(vicky, 0)
plt.imshow(vetim)
```

```python
#Flipping the image (-1 - Flips Vertical axis)
vetim1 = cv2.flip(vicky, -1)
plt.imshow(vetim1)
```

```python
#Flipping the image (1 - Flips Vertical axis)
vetim1 = cv2.flip(vicky, 1)
plt.imshow(vetim1)
```

### Drawing on Images

```python
# Create a blank Canva
np.zeros(3) #it normally gives a float
```

```python
blank = np.zeros(shape=(600,600,3), dtype=np.int16) 
# we add the dtype to make it an int
# you dont have to use, 600,600 you can use any number
# remember its width, height and color channel
```

```python
vicky
```

```python
plt.imshow(blank)
```

```python
# to draw, you need to identify the Top Left Corner  TLC & Bottom Right Corner
```

```python
# draw a rectangle; pt1(TLC - Assumed to be 200,200) pt2(BRC - 400,400)
cv2.rectangle(blank, pt1=(200,200), pt2=(400,400), color=(255,0,0), thickness=10)
plt.imshow(blank)
```

```python
# draw another rectangle; pt1(TLC - Assumed to be 250,250) pt2(BRC - 350,350)
cv2.rectangle(blank, pt1=(200,200), pt2=(400,400), color=(255,0,0), thickness=10)
cv2.rectangle(blank, pt1=(250,250), pt2=(350,350), color=(0,255,0), thickness=3)
plt.imshow(blank)
```

```python
#Drawing a Circle
cv2.circle(blank, center=(300,300), radius=30, color=(0,0,255), thickness=5)
plt.imshow(blank)

# 2 important things are the center and radius
```

```python
#Drawing a Circle
cv2.circle(blank, center=(300,300), radius=200, color=(0,0,255), thickness=5)
plt.imshow(blank)

# 2 important things are the center and radius
```

```python
#Drawing a Circle
cv2.circle(blank, center=(500,50), radius=40, color=(255,255,255), thickness=-1)
plt.imshow(blank)

# 2 important things are the center and radius
```
